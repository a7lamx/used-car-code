```python
Q1 = data1.car_model_year.quantile(0.25)
Q3 = data1.car_model_year.quantile(0.75)
Q1,Q3
```

```python
IQR = Q3 - Q1 
IQR
```

```python
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
lower_limit , upper_limit
```
